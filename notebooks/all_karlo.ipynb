```python
# calculating percentages of missing values
def missing_values_percentages(df: pd.DataFrame):
    """Calculate the percentage of missing values in each column of a DataFrame.
    Args:
        df (pd.DataFrame): The DataFrame to analyze.
    Returns:
        pd.DataFrame: A DataFrame containing the percentage of missing values for each column.
    """
    missing_percent = (df.isna().sum() /len(df)) * 100
    missing_df = pd.DataFrame({'Column': df.columns, "Missing (%)": missing_percent})
    missing_df = missing_df[missing_df["Missing (%)"] > 0].sort_values(by='Missing (%)', ascending=False)

    return missing_df
```

```python
# check missing values percentages
missing_values_percentages(karlo_df)
```

##### Handling of missing values

##### Dropping columns that have the percentage of missing values greater than a given threshold

```python
# dropping columns based on a missing value threshold

def drop_columns_by_missing_threshold(df: pd.DataFrame, threshold: float) -> pd.DataFrame:
    """
    Drops columns from a DataFrame where the percentage of missing values exceeds the given threshold.
    
    Args:
        df (pd.DataFrame): The DataFrame to analyze.
        threshold (float): The threshold percentage (e.g., 50 for 50%).
    
    Returns:
        pd.DataFrame: The DataFrame with columns dropped based on the threshold.
    """
    missing_percent = (df.isna().sum() / len(df)) * 100
    columns_to_drop = missing_percent[missing_percent > threshold].index
    df = df.drop(columns=columns_to_drop)
    
    return df
```

```python
# imputing the missing values with the mode for numerical columns and desriptive key word for object columns
def impute_missing_values(df: pd.DataFrame) -> pd.DataFrame:
    """
    Imputes missing values in the DataFrame.
    
    Numerical columns are filled with the mode, and object columns are filled with a descriptive keyword.
    
    Args:
        df (pd.DataFrame): The DataFrame to process.
    
    Returns:
        pd.DataFrame: The DataFrame with missing values imputed.
    """
    for column in df.columns:
        if df[column].dtype == 'object':
            df[column].fillna('Missing', inplace=True)
        else:
            df[column].fillna(df[column].mode()[0], inplace=True)
    
    return df
```

```python
def check_outliers(df):
    # List to store outlier details for each column
    outlier_data = []

    # Loop through numeric columns
    for col in df.select_dtypes(include=['float64', 'int64']).columns:
        # Compute Q1, Q3, and IQR
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        # Compute outlier boundaries
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Count outliers
        outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
        outlier_count = outliers.shape[0]
        outlier_percentage = (outlier_count / df.shape[0]) * 100

        # Append details to the list
        outlier_data.append({
            'column': col,
            'outlier_count': outlier_count,
            'outlier_percentage': outlier_percentage,
            'lower_bound': lower_bound,
            'upper_bound': upper_bound
        })

    # Convert list to DataFrame
    outlier_df = pd.DataFrame(outlier_data)
    return outlier_df
```
